# Finite differences for approximating derivatives

## Libraries

```python
import numpy as np
from numpy import exp,sin,cos,pi
import matplotlib.pyplot as plt
import pandas as pd
```

## Stepsize

```python
h = np.array([pow(10,-k) for k in range(16)])
```

## $f(x)=\exp(x)$, $x_0=1$

```python
def f(x):
    return exp(x)

def fp(x):
    return exp(x)

def fpp(x):
    return exp(x)

x0 = 1
f0 = f(x0)
fp0 = fp(x0)
fpp0 = fpp(x0)

x0,f0,fp0,fpp0
```

### Forward differences

```python
fwd = (f(x0+h)-f0)/h
eabs = abs(fwd - fp0)
erel = eabs/abs(fp0)
dfwd = pd.DataFrame(np.column_stack((h,fwd,eabs,erel)),columns = ['h','forward difference','absolute error','relative error'])
dfwd.set_index('h', inplace=True)
dfwd
```

### Backward differences

```python
bwd = (f0-f(x0-h))/h
eabs = abs(bwd - fp0)
erel = eabs/abs(fp0)
dbwd = pd.DataFrame(np.column_stack((h,bwd,eabs,erel)),columns = ['h','backward difference','absolute error','relative error'])
dbwd.set_index('h', inplace=True)
dbwd
```

### Centered differences

```python
cen = (f(x0+h)-f(x0-h))/2/h
eabs = abs(cen - fp0)
erel = eabs/abs(fp0)
dcen = pd.DataFrame(np.column_stack((h,cen,eabs,erel)),columns = ['h','centered difference','absolute error','relative error'])
dcen.set_index('h', inplace=True)
dcen
```

### Centered differences for second derivative

```python
two = (f(x0+h)-2*f0+f(x0-h))/h/h
eabs = abs(two - fpp0)
erel = eabs/abs(fpp0)
dtwo = pd.DataFrame(np.column_stack((h,cen,eabs,erel)),columns = ['h','centered difference (order 2)','absolute error','relative error'])
dtwo.set_index('h', inplace=True)
dtwo
```

## $f(x)=\sin(x)$, $x_0=\pi/4$

```python
def f(x):
    return sin(x)

def fp(x):
    return cos(x)

def fpp(x):
    return -sin(x)

x0 = pi/4
f0 = f(x0)
fp0 = fp(x0)
fpp0 = fpp(x0)

x0,f0,fp0,fpp0
```
