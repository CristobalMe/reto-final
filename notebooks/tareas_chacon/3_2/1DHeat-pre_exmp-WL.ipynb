# Example a = 2, D = 1

```
f[x_] := Sin[4*Pi*x]
```

```
Plot[f[x], {x, 0, 1},AxesLabel->{x,f}]
```

```

Plot[f[x], {x, 0, 1},AxesLabel->{x,f}]
```

```
usol=DSolve[{D[u[x, t], t]+ 2*D[u[x, t], x] == D[u[x, t], {x, 2}], u[0, t] == 0, 
  u[1, t] == 0, u[x, 0] == f[x]}, u, {x, t}]
```

```
unsol=NDSolve[{D[u[x, t], t] + 2*D[u[x, t], x] == D[u[x, t], {x, 2}], u[0, t] == 0, u[1, t] == 0, 
 u[x, 0] == f[x]}, u, {x, 0, 1}, {t, 0, 1}];
```

```
unsol = u/.unsol[[1]]
```

```
Plot3D[unsol[x, t], {x, 0, 1}, {t, 0, 0.01}, PlotRange -> All, ColorFunction -> "DarkRainbow", BoxRatios -> {1, 1, 1},
Mesh -> None, AxesLabel->{x,t,u}]
```

```
Plot[Evaluate[Table[unsol[x, t], {t, 0, 0.01, 0.002}]], {x, 0, 1}, 
 PlotRange -> All, AxesLabel->{x,u}]
```

```
xmin = 0;
xmax = 1;
tmin = 0;
tmax = 1;
n = 20;
dx = (xmax - xmin)/n;
dt = 0.001;
m = (tmax - tmin)/dt;
```

```
uexmp=Table[Table[unsol[i dx, k dt], {i, 0, n}], {k, 0, m}];
Export["uexmp_1_2_20.csv", uexmp, "CSV"]
Dimensions[uexmp]
```

```
xmin = 0;
xmax = 1;
tmin = 0;
tmax = 1;
n = 40;
dx = (xmax - xmin)/n;
dt = 0.001;
m = (tmax - tmin)/dt;
```
