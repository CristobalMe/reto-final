## $t\in [0, 0.1], \Delta t = .001, \Delta x = 0.1, s = 0.1$

```
unsol=NDSolve[{D[u[x, t], t] == D[u[x, t], {x, 2}], u[0, t] == g0[t], 
  u[1, t] == g1[t], 
 u[x, 0] == f[x]}, u, {x, 0, 1}, {t, 0, 0.1}];
```

```
unsol = u/.unsol[[1]]
```

```
Plot3D[unsol[x, t], {x, 0, 1}, {t, 0, 0.1}, PlotRange -> All, ColorFunction -> "DarkRainbow", BoxRatios -> {1, 1, 1},
Mesh -> None, AxesLabel->{x,t,u}]
```

```
Plot[Evaluate[Table[unsol[x, t], {t, 0, 0.1, 0.02}]], {x, 0, 1}, 
 PlotRange -> All, AxesLabel->{x,u}]
```

```
xmin = 0;
xmax = 1;
tmin = 0;
tmax = 0.1;
dx = 0.01;
n = (xmax - xmin)/dx;
dt = 0.0001;
m = (tmax - tmin)/dt;
```

```
uexmp=Table[Table[unsol[i dx, k dt], {i, 0, n}], {k, 0, m}];
Export["uexmpdnh.csv", uexmp, "CSV"]
Dimensions[uexmp]
```
